```python
# Filtrar los datos según la condición slope = 0
data = df[df['slope'] == 0]
data1 = df[df['slope'] == 1]
data2 = df[df['slope'] == 2]

# Calcular el porcentaje de cada valor de 'target' dentro de los datos filtrados
counts = data['target'].value_counts(normalize=True) * 100
counts1 = data1['target'].value_counts(normalize=True) * 100
counts2 = data2['target'].value_counts(normalize=True) * 100

fig, (ax1, ax2) = plt.subplots(2, 2, figsize=(10, 10))

# Convertir los porcentajes y las etiquetas en un DataFrame
df_counts = pd.DataFrame({'Porcentaje': counts.values}, index=counts.index)
df_counts1 = pd.DataFrame({'Porcentaje': counts1.values}, index=counts1.index)
df_counts2 = pd.DataFrame({'Porcentaje': counts2.values}, index=counts2.index)

# Graficar el diagrama de pie con la biblioteca Seaborn en el primer subplot
sns.set()
#sns.color_palette("bright")

ax1[0].set_title('Porcentaje de target cuando slope = 0')
ax1[0].pie(df_counts['Porcentaje'], labels=df_counts.index, autopct='%1.1f%%', startangle=90)
ax1[0].axis('equal')


ax1[1].set_title('Porcentaje de target cuando slope = 1')
ax1[1].pie(df_counts1['Porcentaje'], labels=df_counts1.index, autopct='%1.1f%%', startangle=90, colors=[ '#DD8452', '#4C72B0'])
ax1[1].axis('equal')

ax2[0].set_title('Porcentaje de target cuando slope = 2')
ax2[0].pie(df_counts2['Porcentaje'], labels=df_counts2.index, autopct='%1.1f%%', startangle=90, colors=[ '#DD8452', '#4C72B0'])
ax2[0].axis('equal')


plt.show()
```

```python
# Crear una figura y un conjunto de ejes
corr_matrix = df.corr()
fig, ax = plt.subplots(figsize=(25,25))

# Aumentar el tamaño de la fuente
sns.set(font_scale=2)

# Graficar el heatmap con los datos y las opciones
sns.heatmap(corr_matrix, annot=True, cmap="YlGnBu")

# Mostrar el gráfico
plt.show()
```

## Análisis de enfermedades cardíacas

En este estudio se analizó un dataset de 270 personas con edades entre 30 y 77 años para determinar factores de riesgo y patrones en la incidencia de enfermedades cardíacas.

Se encontró que el 45.4% de los hombres en el estudio presentaban alguna enfermedad cardíaca, en comparación con solo el 23% de las mujeres.

Además, se observó que gran parte de los casos en hombres estaban entre los 50 y 60 años de edad, mientras que en mujeres los casos se encontraban entre los 57 y 64 años.

Entre los factores de riesgo identificados se encontró que la angina inducida por el ejercicio es un indicador de alguna enfermedad cardíaca. De igual manera, se descubrió que un segmento ST durante el ejercicio con una pendiente plana o descendente también podría ser un indicador de enfermedad cardíaca.

Finalmente, se identificó que presentar talasemia podría ser un factor de riesgo adicional para enfermedades cardíacas.

En resumen, estos hallazgos pueden ser útiles para la prevención y el tratamiento de enfermedades cardíacas en diferentes grupos de edad y género.
